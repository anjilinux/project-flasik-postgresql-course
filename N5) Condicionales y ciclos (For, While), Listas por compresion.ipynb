$Edad_{Humana}= 22 +(Edad_{Perro}-2)*5 , Edad_{Perro}\geq 2$

```python
age = int(input("Ingresa la edad de tu perro: "))

if age < 0:
    print("No puede ser cierto!")
elif age == 0:
    print("Corresponde a la edad humana de 0 años!")
elif age == 1:
    print("Aproximadamente 14 años!")
elif age == 2:
    print("Aproximadamente 22 años!")
else:
    human = 22 + (age -2) * 5
    print("Corresponde a  " + str(human) + " años humanos!")
```

$BMI=\frac{M}{I^2}$

Donde :
M= Masa en Kg
I= estatura en m

```python
height = float(input("Cual es tu altura? "))
weight = float(input("cual es tu peso? "))

bmi = weight / height**2
print('BMI:',bmi)
if bmi < 15:
    print("Extremo bajo peso")
elif bmi < 16:
    print("Severo bajo peso")
elif bmi < 18.5:
    print("Bajo peso")
elif bmi < 25:
    print("Normal (saludable)")
elif bmi < 30:
    print("Sobre peso")
elif bmi < 35:
    print("Obesidad Clase I (Moderada)")
elif bmi < 40:
    print("Obesidad Clase II (Severa)")
else:
    print("Obesidad Clase III (Muy severa)")
```

$a^2 +b^2 =c^2$

```python
from math import sqrt
n = int(input("Numero maximo? "))
for a in range(1, n+1):
    for b in range(a, n):
        c_square = a**2 + b**2
        c = int(sqrt(c_square))
        if ((c_square - c**2) == 0):
            print(a, b, c)
```

$\color{blue}{\text{Si recorre una lista, es mejor evitar cambiar la lista en el cuerpo del bucle. Eche un vistazo al siguiente ejemplo:}}$

Para evitar estos efectos secundarios, es mejor trabajar en una copia utilizando el operador interno, como se puede ver en el siguiente ejemplo:

```python
# Iteraciones con efectos
colours = ["red"]
for i in colours:
    if i == "red":
        colours += ["black"]
    elif i == "black":
        colours += ["white"]
print(colours)
```

```python
colours = ["red"]
for i in colours[:]:
    if i == "red":
        colours += ["black"]
    elif i == "black":
        colours += ["white"]
print(colours)
```
