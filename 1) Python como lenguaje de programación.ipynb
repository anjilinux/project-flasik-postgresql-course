2. Se usa para desarrollo web, desarrollos matematicos y cientificos, desarrollo de aplicaciones entre otras.
3. Provee un excelente soprte de librerías con una gran cantidad de desarrolladores.

# Caracteristicas de Python 

plot_3.jpg

1. **Simplicidad:** Es considerado un lenguaje muy fácil de aprender y entender
2. **Plataforma independiente:** Python puede correr de la misma forma en diversas plataformas como Windows, Linux y Unix y muchas mas. Mismo codigo funciona en cualquier sistema operativo.
3. **Libre acceso:** Usando internet se puede descargar facilmente y sin problemas
4. **Lenguaje de alto nivel:** No es necesario de utilizar la memoria 
5. **Interpreter:** Es una condicion que indica que se ejecuta linea a linea haciendo el proceso de ejecucion simple 
6. **Soporte de librerias:** Existe una gran variedad y soporte de librerias, por ende se facilitan los procesos ya que existen muchos problemas ya resueltos
7. **Propiedades Embedding:** Puede ser incoporado codigo de otros lenguajes lo que lo hace muy potente
8. **Orientado a objetos:** Esto provee alta flexibilidad para los desarrolladores en el desarrollo de proyectos.
