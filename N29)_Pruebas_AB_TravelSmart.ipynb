## Introduccion

**Contexto empresarial.** Su empresa, TravelSmart, es una empresa en crecimiento y compite con algunos grandes actores del mercado. Como parte de una promoción de marketing para aumentar la base de usuarios, ha decidido ofrecer un descuento adicional del 5 % a los usuarios que reserven vuelos y hoteles en TravelSmart.com, en lugar de solo billetes de avión. Le interesa saber cómo este acuerdo de marketing afecta una variedad de otras métricas, como vistas de página, usuarios, sesiones, ingresos, etc. en comparación con el estado actual.
     
**Problema comercial.** La empresa desea que responda la siguiente pregunta: **"¿Deberíamos ofrecer esta promoción de descuento del 5 % si nuestro objetivo es aumentar el número de clientes que compran?"**

**Contexto analítico.** El escenario anterior puede verse como una **prueba A/B**. Las pruebas A/B son una forma de determinar si un cambio potencial ("Versión B", a diferencia de la "Versión A" existente) mejora las métricas comerciales clave, como las ventas o las conversiones.

A-B.png
     
El caso está estructurado de la siguiente manera:  (1) definiras y estableceras una hipótesis precisa; (2) implementar la prueba A/B; (3) analizar los resultados de la prueba; y finalmente (4) decidir sistemáticamente los próximos pasos de si debe ir e implementar un cambio o quedarse con la versión actual.

## Diseñando el experimento: desarrollando una hipótesis concisa

Los espaguetis son deliciosos, pero aún más deliciosos cuando la pasta está bien cocida. Muchos chefs usan la "prueba de los espaguetis": tiran los espaguetis a la pared y, si se pegan, están cocidos. Este es un gran ejemplo de diseño de hipótesis deficiente porque no da ninguna indicación de cuándo o por qué se debe considerar hecho el espagueti.

Al igual que cocinar espaguetis, antes de proceder con cualquier experimento o prueba (en este caso, una prueba A/B), necesitamos definir una hipótesis muy precisa para que tengamos un contexto sobre el cual evaluar los resultados de nuestro experimento o prueba.

## Identificar variables control

Ahora que tenemos una hipótesis concisa, debemos pensar qué podría salir mal y cómo podemos prevenir/mitigar problemas potenciales. Un aspecto importante a considerar son las **variables de control**. Las variables de control son elementos cuantificables que permanecen sin cambios a lo largo de un experimento. Identificar las variables de control es muy importante para mitigar los datos ruidosos porque las ligeras variaciones en las variables que espera que permanezcan constantes a lo largo de su experimento pueden afectar en gran medida los resultados que se miden, de maneras que no tienen nada que ver con lo que le interesa probar específicamente.

Por ejemplo, digamos que Taylor se pone a dieta y aumenta el consumo de verduras durante las próximas semanas. Taylor quiere saber si esto conducirá a la pérdida de peso. Taylor regularmente sale a caminar con su perro todas las mañanas. Durante las próximas semanas, Taylor siente que le falta energía para levantarse por la mañana para su caminata diaria y, a menudo, no sale a caminar. Taylor termina ganando unos kilos.

Este es un ejemplo de un experimento mal diseñado. Los pocos kilos de más podrían deberse a varias cosas, incluida la falta de ejercicio. Por lo tanto, no es posible saber si el consumo de vegetales contribuyó a la pérdida de peso a partir de este experimento. Si Taylor identificara el ejercicio diario como una variable de control, este experimento habría tenido mucho más éxito.

## Definiendo KPIs (Key Performance Metrics)

Una vez que hayamos identificado las hipótesis y variables relevantes, estamos casi listos para implementar nuestro experimento. Lo último que debemos considerar es elegir qué variable debemos medir para que la evaluación del resultado de este experimento sea muy clara. Estas variables se denominan **Indicadores clave de rendimiento (KPI)**. Hay muchas métricas diferentes que podemos considerar, pero todas se pueden clasificar como **micro o macro conversiones**.

Las **macroconversiones** son las conversiones principales para un servicio en particular. En este caso, las macroconversiones se refieren a las *métricas de acumulación de clientes del sitio web*. 

Las **microconversiones** se relacionan con los componentes más pequeños que se suman a las macroconversiones. La siguiente tabla son algunos ejemplos de macro y micro conversiones para TravelSmart:

| Macro Conversions            | Micro Conversions           |
| ---------------------------- | --------------------------- |
| Número total de ventas de boletos | Vistas en una ruta específica   |
| Compras de membresía         | Tiempo dedicado a la página de destino  |
| Conversiones de ventas y clientes potenciales  | Número de cuentas de refereridos |

Hagamos el siguiente experimento para comprender mejor las conversiones macro/micro. Supongamos que ejecutó su oferta durante 60 días, y en el día 60 llegamos a algunos datos proporcionados en el siguiente gráfico:

KPI.png

La siguiente tabla resume el gráfico anterior. Analicemos lo que está pasando en este escenario ficticio.

```python
def color_negative_red(value):
    if '%' not in str(value):
        return ''
    value = float(value.rstrip('%'))
    if value < 0:
        color = '#d65f5f'
    elif value > 0:
        color = '#5fba7d'
    else:
        color = 'black'
    return 'background-color: %s' % color
```

```python
days = 60
df_experiment = pd.read_csv('https://raw.githubusercontent.com/dfbustosus/Curso_DS_para_todos/main/Data/micro_macro_dataset.csv')
df_experiment
```

```python
# Elegir los 60 dias anteriores
pre_experiment = df_experiment.loc[1:days]
# Elegir los 120 dias posteriores (aunque solo hay 60)
post_experiment = df_experiment.loc[days+1:2*days]
```

```python
pre_means = pre_experiment.mean().apply(floor) # entero mas cercano abajo
pre_means
```

```python
post_means = post_experiment.mean().apply(floor)
post_means
```

```python
percentage_changes = ((post_means - pre_means)/pre_means*100).map("{:.2f}%".format)
```

```python
means = pd.DataFrame({
    "Pre-Experiment Mean": pre_means,
    "Post-Experiment Mean": post_means,
    "Percentage Changes" : percentage_changes
}).transpose()
means
```

```python
means.style.applymap(color_negative_red)
```

Aunque vemos que hay más visitas a estas páginas específicas, la cantidad de boletos comprados y hoteles reservados no cambió significativamente. Esto se refleja negativamente en nuestro experimento porque parece que no hubo cambios en la venta de boletos, pero en realidad perdimos dinero al ofrecer precios más bajos de boletos y hoteles. En este caso, decimos que las macroconversiones no se vieron afectadas por este experimento y consideramos que el experimento no tuvo éxito.

Nuestra conclusión principal es que cada vez que definimos un experimento, es importante definir y realizar un seguimiento de las macroconversiones como nuestros indicadores clave de rendimiento. Las microconversiones también son importantes, pero el enfoque principal debe estar en las macroconversiones.

Para nuestro experimento, la macroconversión más sensata para usar como KPI es la cantidad de usuarios distintos que compraron un boleto o reservaron un hotel, porque esto está directamente relacionado con nuestra base de usuarios, que es exactamente lo que estamos tratando de expandir.

## Tratamiento y grupos de control

Muy bien, ahora estamos listos para implementar nuestro experimento. Como implica el nombre de esta prueba, tenemos que dividir nuestra base de usuarios en dos grupos, los grupos A y B. De esa manera, podemos aplicar el experimento en el grupo B y ver cómo se compara con el grupo sin cambios, el grupo A. Nos referimos al grupo B como el grupo de tratamiento (el "tratamiento" aquí se refiere al trato que se ofrece) y al grupo A como el grupo de control.

## Aumentando el experimento

Ahora que ha ejecutado su experimento para el 2% de la base de usuarios durante un par de días, es el momento de aumentar su experimento. Antes de discutir cómo y cuándo aumentar, analicemos primero cómo comprender los resultados de un experimento A/B.

Supongamos que decide otorgar aleatoriamente la oferta al 10 % de los usuarios y la ejecuta durante un par de días. Si está interesado en comprar usuarios, puede calcular la diferencia en el porcentaje de usuarios de compras entre los dos grupos. Supongamos que ejecutamos el experimento y observamos el siguiente resultado para la diferencia en el porcentaje de usuarios que compran entre el grupo A y el grupo B:

El intervalo de confianza del 95% indica que existe una probabilidad del 95% de que el porcentaje de usuarios compradores del grupo B sea entre un 0,7% y un 1,2% mayor que el porcentaje de usuarios compradores del grupo A. En este caso, el 0% está fuera del intervalo, y podemos concluir que el porcentaje de usuarios que compran es mayor para los usuarios que reciben el trato. Además, el valor $p$ - de 0,03 indica que este resultado es significativo.

Si el experimento se ejecuta durante más tiempo, el intervalo de confianza será más estrecho. Esto se debe a que la cantidad de usuarios a los que se otorga el trato aumenta con el tiempo. Este aumento de información conduce a una mayor precisión en nuestra estimación.

## Para recordar

Las pruebas A/B son una herramienta poderosa para realizar una experimentación rápida y aprender rápidamente sobre las preferencias de sus usuarios. Sin embargo, cualquier prueba A/B que realice debe garantizar una aleatorización adecuada. Garantiza que no excluirá accidentalmente ninguna base de usuarios y sacará conclusiones basadas en una muestra sesgada. Con una prueba A/B, generalmente se inicia el experimento en un pequeño porcentaje de usuarios para mitigar el riesgo. Después de este paso inicial, el experimento puede escalarse.
