# Importancia del return 
La declaración return [expresión] sale de una función y, opcionalmente, devuelve una expresión. Una declaración return sin argumentos es lo mismo que return None.

```python
# cuando retorno
def reverse(lista):
    lista.reverse()
    return lista

# cuando no retorno
def reverse2(lista):
    lista.reverse()
    print(lista)
```

```python
l = [1,2,3]
a = reverse(l)
print(a)
```

```python
l = [1,2,3]
a=reverse2(l)
print(a)
```

# Variables locales vs globales

Las variables que se definen dentro de un cuerpo de función tienen un alcance local y las definidas fuera tienen un alcance global.

Esto significa que solo se puede acceder a las variables locales dentro de la función en la que están declaradas, mientras que todas las funciones pueden acceder a las variables globales en todo el cuerpo del programa. 

Las variables locales se crean cuando la función ha comenzado la ejecución y se pierden cuando la función termina, por otro lado, la variable global se crea cuando comienza la ejecución y se pierde cuando finaliza el programa.

```python
total = 0; # Variable global
# Definicion de funcion
def sum1( arg1, arg2 ):
   # Adicionar dos valores numericos."
   total = arg1 + arg2; # Esta es la variable local
   print("Dentro de la funcion variable total total : ", total)
   return total;

#Usando la funcion
sum1( arg1=10, arg2=20 );
print("Afuera de la funcion variable global total : ", total )
```

```python
total = 123; # Variable global
# Definicion de funcion
def sum2( arg1, arg2 ):
  '''
  input:
  arg1: es un numero
  arg2: es numero
  output:
  total= suma de arg1 y arg2
  prod= multiplicacion de arg1 y arg2
  div= division arg1 y arg2
  '''
  # Adicionar dos valores numericos."
  total = arg1 + arg2; # Esta es la variable local
  prod= arg1*arg2
  div=arg1/arg2
  return total, prod, div;
```

# Recursividad

En Python, sabemos que una función puede llamar a otras funciones. Incluso es posible que la función se llame a sí misma. Estos tipos de construcciones se denominan **funciones recursivas**.

```python
# iterador contador y condicion

count = 0
def recorrer(count):
    # print('Inicio recursividad')
    print(count)
    if count < 10:
        count += 1
        recorrer(count)
    else:
        print('fin')
```

```python
recorrer(count=3)
```

```python
def factorial(x):
    if x == 1:
        return 1
    else:
        return (x * factorial(x-1))
num = 3
print("El factorial del numero", num, "es", factorial(num))
```

# Secuencia de Fibonacci

$a_n= F_{n-1} +F_{n-2}$

Cada numero es la suma de sos dos anteriores

```python
def fibo(pos):
    if pos<1:
        print('error')
    elif pos == 1:
        return 0
    elif pos == 2:
        return 1
    else:
        return fibo(pos - 1) + fibo(pos - 2)
        
```

```python
fibo(5) # 0,1,1,2,3,5,8,13, # Trigger
```

```python
for i in range(1,10):
    print(fibo(i))
```

# Funciones Especiales

El principal uso de __args__ y __kwargs__ es en la definición de funciones. Ambos permiten pasar un número variable de argumentos a una función, por lo que si quieres definir una función cuyo número de parámetros de entrada puede ser variable, considera el uso de *args o **kwargs como una opción

ejemplo.jpg

```python
def op_functions(l,*args):
    
    if args:
        print('hay argumentos extra')
        print('Argumento extra: ',args)
        if args[0] == 1:
            '''
                1 = Promedio
            '''
            prom = 0
            for i in l:
                prom += i
            return prom/len(l)
        elif args[0] == 2:
            '''
                2 = Par o impar
            '''
            result = []
            for i in l:
                if i % 2 == 0:
                    result.append('par')
                else:
                    result.append('impar')
            return result
        elif args[0] == 3:
            '''
                3 = Mediana
            '''
            if len(l) % 2 == 0:
                return ((l[len(l)//2]) + (l[(len(l)//2) - 1]))/2
            else:
                #return f'Mediana: {l[len(l)//2]}'
                return l[len(l)//2]
    else:
        print('no hay argumentos extra')
```

```python
lista = [1,2,3,4,5]
```

```python
op_functions(lista,1)
```

```python
op_functions(lista,2)
```

```python
op_functions(lista,3)
```

```python
op_functions(lista)
```

**Y puedo usar args y kargs**

```python
def op_functions_2(l,**kargs):
    
    if kargs.keys():
        print('hay argumentos extra')
        # print(kargs.keys())
        if 'prom' in kargs.keys():            
            if kargs['prom'] == True:
                '''
                    Promedio
                '''
                prom = 0
                for i in l:
                    prom += i
                return prom/len(l)
        
        if 'odd' in kargs.keys():
            
            if kargs['odd'] == True:
                '''
                    Par o impar
                '''
                result = []
                for i in l:
                    if i % 2 == 0:
                        result.append('par')
                    else:
                        result.append('impar')
                return result
            
        if 'median' in kargs.keys():
            
            if kargs['median'] == True:
                '''
                    Mediana
                '''
                if len(l) % 2 == 0:
                    return ((l[len(l)//2]) + (l[(len(l)//2) - 1]))/2
                else:
                    return l[len(l)//2]
    else:
        print('no hay argumentos extra')
```

```python
op_functions_2(lista)
```

```python
op_functions_2(lista, prom=True)
```

```python
op_functions_2(lista, odd=True)
```

```python
op_functions_2(lista, median=True)
```

```python
def op_functions_3(l, *args, **kargs):
    
    '''
        args disponibles:
        1 = prom
        2 = par o impar
        3 = mediana
    '''
    
    '''
        kargs disponibles:
        'prom' == True => retorna el promedio
        'odd' == True => retorna una lista con los pares e impares
        'meadian' == True => retorna la mediana de la lista
    '''
    
    if args:
        print('hay parametros simples adicionales')
        print(args)
    
    if kargs.keys():
        print('hay parametros complejos adicionales')
        print(kargs)
    
```

```python
op_functions_3(lista, 1, prom=True)
```
