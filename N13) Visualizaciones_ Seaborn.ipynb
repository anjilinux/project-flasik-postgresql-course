```python
import seaborn as sns
import pandas as pd
import numpy as np
```

imagen.png

Las funciones **tipo Axis** trazan datos en un solo objeto matplotlib.pyplot.Axes, que es el valor de retorno de la función.

Por el contrario, las funciones a **nivel figure** interactúan con matplotlib a través de un objeto llamado FacetGrid, que administra la figura. Cada módulo tiene una sola función a nivel de figura, que ofrece una interfaz unitaria a sus diversas funciones axis.

```python
url='https://raw.githubusercontent.com/skalsangra/Bank-Marketing-Data-Analysis/master/marketing-data.csv'
df=pd.read_csv(url,sep=',')
df.head()
```

# Figure level

## relplot

```python
sns.relplot?
```

```python
ax=sns.relplot(x='age', y='balance', kind='scatter',hue='education',data=df)
ax.tight_layout()
ax.set_xlabels(label='Edad')
ax.set_ylabels(label='Balance')
ax.set_titles(label='Balance vs Edad')
```

```python
x = np.linspace(0, 10, 100)
y= np.cos(x)
ax=sns.relplot(x,y, kind='line', color='red')
ax.tight_layout()
```

## displot

```python
sns.distplot?
```

```python
ax=sns.distplot(df.age,bins=20,hist=True, kde=False)
```

```python
ax=sns.distplot(df.age,bins=20,hist=False, kde=True)
```

## catplot

```python
sns.catplot?
```

```python
ax=sns.catplot(x='marital',y='balance',data=df,alpha=0.5,kind='strip')
```

```python
ax=sns.catplot(x='marital',y='balance',data=df,kind='box',showfliers=False)
```

```python
ax=sns.catplot(x='marital',y='balance',data=df,kind='violin')
```

```python
ax=sns.catplot(x='marital',y='balance',data=df,kind='boxen',showfliers=False)
```

```python
df.groupby('marital').describe()['balance']
```

```python
# devuelve IC
ax=sns.catplot(x='marital',y='balance',data=df,kind='point',showfliers=False)
```

```python
ax=sns.catplot(x='marital',y='balance',data=df,row='loan',col='housing',kind='box',showfliers=False)
```

# Axis-level

## scatterplot

```python
sns.scatterplot?
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sns.scatterplot(x='age', y='balance',hue='education',data=df)
```

## lineplot

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
x = np.linspace(0, 10, 100)
y= np.cos(x)
sns.lineplot(x,y, color='red')
```

## histplot

```python
plt.figure(figsize=(12,6))
sns.histplot(df.age,bins=20)
```

##kdeplot

```python
plt.figure(figsize=(12,6))
sns.kdeplot(df.age,fill=True)
```

## ecdfplot

```python
plt.figure(figsize=(12,6))
sns.ecdfplot(df.age)
```

##striplot

```python
plt.figure(figsize=(12,6))
sns.stripplot(x='marital',y='balance',data=df,alpha=0.5)
```

## boxplot

```python
sns.boxplot(x='marital',y='balance',data=df,showfliers=False)
```

## violinplot

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='marital',y='balance',hue='education',data=df,showfliers=False)
plt.legend()
```

## barplot

```python
t=pd.DataFrame(df[['marital','education']].value_counts(),columns=['conteo'])
t=t.reset_index()
t
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='marital',y='conteo',hue='education',data=t)
```
