```python
import math
print(format(math.pi, '.2f'))
print(format(math.pi, '.3f'))
a=round(math.pi,3)
print(a)
print(math.ceil(a))
print(math.floor(a))
```

```python
# Numeros complejos
c= 3-7j
print(type(c))
print(c.real)
print(c.imag)
```

```python
# Operaciones
print(2.5+2.1)
print(4-3.5)
print(5*4)
print(5/6)
print(4%3) # Modulo 
print(5//2) # Floor division
2**2 # Exponenciacion ^
```

```python
print(abs(-2.5))
print(int(2.5))
```

```python
print(abs(-2.5)) # Valor absoluto
print(int(2.5))  # Extraer el entero del float
print(float(2.5)) 
print(complex(4,3j)) # La parte compleja le hace set up a 0
c = 3+4j
print(c.conjugate())
print(divmod(94, 21)) # (x // y, x % y)
print(pow(3, 2))
x = 123.56
print(math.trunc(x)) # Truncar a enteros
x = 123.56
print(round(x, 1))
```

```python
# String Formatting Operator (el simbolo % me permite guardar variables dentro de un print)
print("Hola %s, has obtenido un score de %i del total de %f" % ("David", 3, 100))
print("Hola %s, has obtenido un score de %f del total de %f" % ("Juan", 2.15, 2.5))
```
